### Correlation Plot using matplotlib

```python
# Correlation Heatmap using matplotlib
plt.matshow(df_trim.corr())
plt.show()
```

```python
# Create figure with specified size
f = plt.figure(figsize=(10, 8))  # Increased figure width

# Plot correlation matrix as heatmap
plt.matshow(df_trim.corr(), fignum=f.number)

# Customize x-axis labels
labels = df_trim.columns.tolist()
wrapped_labels = [textwrap.fill(label, 10) for label in labels]
plt.xticks(range(len(labels)), wrapped_labels, rotation=90, fontsize=8, ha='right')

# Customize y-axis labels
plt.yticks(range(df_trim.shape[1]), df_trim.columns, fontsize=8)

# Add colorbar with font size adjustment
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)

# Add title with font size adjustment
plt.title('Correlation Matrix', fontsize=14)

# Adjust subplot parameters
plt.subplots_adjust(
    left=0.2,  # Make room for y-axis labels
    bottom=0.3,  # Make room for x-axis labels
    right=0.8,  # Adjust right margin
    top=0.9  # Adjust top margin
)

# Save plot with tight bounding box and DPI
plt.savefig('plot.png', bbox_inches='tight', dpi=300)

plt.show()
```

```python
f.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'correlation.png'))
```

### Correlation plot using seaborn

```python
# Create subplot with customized size
f, ax = plt.subplots(figsize=(10, 10))

# Plot correlation matrix as heatmap
labels = df_trim.columns.tolist()
wrapped_labels = [textwrap.fill(label, 10) for label in labels]

# Create correlation heatmap
corr = sns.heatmap(
    df_trim.corr(), 
    annot=True,
    ax=ax, 
    square=True,  # Ensure square cells
    fmt='.2f',  # Format annotation values
    linewidths=0.5  # Cell border width
)

# Customize x-axis labels
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(wrapped_labels, rotation=90, fontsize=8, ha='right')

# Customize y-axis labels
ax.set_yticks(range(len(labels)))
ax.set_yticklabels(labels, fontsize=8)

# Add title
ax.set_title('Correlation Matrix Heatmap', fontsize=14, fontweight='bold')

# Show plot
plt.tight_layout()
plt.show()
```

```python
f,ax.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'seaborn_corr.png'))
```

### Categorical Plots:

### Creating a categorical variable that splits the "Payment installments" column into categories

```python
# Use a histogram to visualize the distribution of the variables. 
# This way, you can determine sensible categories for the product category. 
# You don't want to end up with too few observations in any one of the categories.
# The argument "kde" add a line that encompasses the distribution

category = sns.histplot(dev_trim_df['payment_installments'], bins = 20, kde = True)
```

```python
# highest buyers pay in just 2 installments
```

```python
# Save plot
category.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'category.png'))
```

```python
dev_trim_df.loc[dev_trim_df['payment_installments'] < 5, 'installment category'] = 'Low'
```

```python
dev_trim_df.loc[(dev_trim_df['payment_installments'] >= 5 ) & (dev_trim_df['payment_installments'] < 10), 'installment category'] = 'Middle'
```

```python
dev_trim_df.loc[dev_trim_df['payment_installments'] > 10, 'installment category'] = 'High'
```

```python
dev_trim_df['installment category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the installment categories created above

sns.set(style="ticks")
category_installments = sns.catplot(x="payment_installments", y="payment_value", hue="installment category", data=dev_trim_df)
```

```python
# Create a categorical plot in seaborn using the installment categories created above

sns.set(style="ticks")
Price_installments = sns.catplot(x="payment_installments", y="price", hue="installment category", data=dev_trim_df)
```

```python
# Payment installments are less if the price of the product is higher. 
# Whereas high number of installments are for the product price which is on the lower side.
```

```python
# Save plot
category_installments.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'category_installments.png'))
```
